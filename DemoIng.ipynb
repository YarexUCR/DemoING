```python
from ultralytics import YOLO
import cv2
import os

# Ruta del modelo YOLOv8 preentrenado
model_path = r'kaggle\working\runs\classify\train\weights\best.pt'

# Verifica si el modelo existe
if not os.path.exists(model_path):
    print(f"El archivo del modelo {model_path} no existe.")
else:
    print(f"El archivo del modelo {model_path} se encontró correctamente.")

# Cargar el modelo YOLOv8
try:
    model = YOLO(model_path)
    print("Modelo cargado exitosamente.")
except Exception as e:
    print(f"Error al cargar el modelo: {e}")
```

```python
# Diccionario para mapear los nombres de las clases en inglés a español
class_names_es = {
    'angry': 'enojado',
    'disgust': 'disgusto',
    'happy': 'feliz',
    'sad': 'triste',
    'surprised': 'sorprendido'
}

# Inicializar la cámara
cap = cv2.VideoCapture(0)  # 0 para la cámara predeterminada

while True:
    # Capturar una imagen de la cámara
    ret, frame = cap.read()
    if not ret:
        break

    # Realizar detección de emociones en la imagen
    results = model.predict(frame)

    # Verificar si se obtuvieron resultados
    if results and hasattr(results[0], 'probs') and results[0].probs is not None:
        # Obtener las probabilidades de las clases
        probs = results[0].probs

        # Obtener el índice de la emoción con mayor probabilidad
        class_id = probs.top1  # No es necesario usar item() aquí
        class_name_en = model.names[class_id]
        class_name_es = class_names_es.get(class_name_en, class_name_en)  # Traducir el nombre de la clase a español
        probability = float(probs.top1conf.item())  # Obtener la probabilidad del top1 como float

        # Mostrar la emoción detectada en la imagen
        cv2.putText(frame, f"Emocion: {class_name_es} ({probability:.2f})", 
                    (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Mostrar la imagen con la emoción detectada
    cv2.imshow('Deteccion de Emociones', frame)

    # Salir del bucle si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
